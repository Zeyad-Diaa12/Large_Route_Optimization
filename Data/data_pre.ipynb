```python
orders_gen = {}
for index, row in orders.iterrows():
    order_id = row['Order_ID']
    if order_id not in orders_gen:
        orders_gen[order_id] = {
            'Source': row['Source'],
            'Destination': row['Destination'],
            'Available_Time': row['Available_Time'],
            'Deadline': row['Deadline'],
            'Items': []
        }
    orders_gen[order_id]['Items'].append({
        'Material_ID': row['Material_ID'],
        'Item_ID': row['Item_ID'],
        'Area': row['Area'],
        'Weight': row['Weight'],
        'Danger_Type': row['Danger_Type']
    })
```

```python
import networkx as nx
import matplotlib.pyplot as plt
# Create a graph
G = nx.Graph()

# Add nodes (cities) to the graph
cities = set(distance['Source']).union(set(distance['Destination']))
G.add_nodes_from(cities)

# Add edges (links between cities) to the graph
for index, row in distance.iterrows():
    G.add_edge(row['Source'], row['Destination'])

# Draw the graph
pos = nx.spring_layout(G)  # Positions for all nodes
nx.draw(G, pos, with_labels=True, node_size=400, node_color="skyblue", font_size=10)
nx.draw_networkx_edges(G, pos, width=0.1, alpha=0.2)
plt.title("Cities and Their Connections")
plt.show()
```

```python
# Create an empty dictionary to store connections
city_connections = {}

# Iterate over the DataFrame and populate the dictionary
for index, row in distance.iterrows():
    if row['Source'] not in city_connections:
        city_connections[row['Source']] = {row['Destination']}
    else:
        city_connections[row['Source']].add(row['Destination'])

    if row['Destination'] not in city_connections:
        city_connections[row['Destination']] = {row['Source']}
    else:
        city_connections[row['Destination']].add(row['Source'])

# Convert sets to lists
city_connections = {city: list(connections) for city, connections in city_connections.items()}
```

```python
import pandas as pd 
import json
distances = pd.read_csv('city_connections_with_time.csv')
```

```python
distances
```

```python
# Create a dictionary to store the distance matrix
distance_matrix = {}

for index, row in distances.iterrows():
    source = str(int(row['Source']))  # Remove decimal and convert to string
    destination = str(int(row['Destination']))  # Remove decimal and convert to string
    distance = row['Distance(KM)']
    time = row['Total_Time(hr)']
    cost = row['Cost_per_KM']
    
    # Add distance, time, and cost to the matrix (both ways if it's symmetric)
    if source not in distance_matrix:
        distance_matrix[source] = {}
    
    distance_matrix[source][destination] = {'distance': distance, 'time': time, 'cost': cost}
```

```python
distance_matrix
```

```python
import json
# Save the dictionary to a JSON file
with open('distance_matrix.json', 'w') as json_file:
    json.dump(distance_matrix, json_file, indent=4)
```
